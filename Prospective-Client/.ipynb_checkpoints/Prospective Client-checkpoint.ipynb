```python
product={}
product['4A']='Health'
product['4B']='Life'
product['4C']='Supplemental'
product['4D']='Dental'
product['4E']='Vision'
product['4F']='Temporary disability'
product['4G']='Prepaid legal'
product['4H']='Long-term disability'
product['4I']='Severance pay'
product['4J']='Apprenticeship and training'
product['4K']='Scholarship (funded)'
product['4L']='Death benefits'
product['4P']='Taft-Hartley Financial Assistance'
product['4Q']='Other'
product['4R']='Unfunded'
product['4S']='Unfunded'
product['4T']='10 or more employer plan'
product['4U']='Collectively-bargained welfare benefit arrangement'
```

```python
new_iowa_Welfare_Benefit_Life_under_200.to_csv('new_iowa_Welfare_Benefit_Life_under_200.csv')
```

```python
def getBenefit(row,productCode):
    obj=(row['TYPE_WELFARE_BNFT_CODE'])
    #print(obj)
    characters=list(obj)
    i=0
    canSell={}
    canSell[productCode]='Yes'
    while i<len(characters):
        x=characters[i]+characters[i+1]
        i=i+2
        try:
            if(product[x] == productCode):
                canSell[productCode]='No'
        except:
            print('error')
    return canSell[productCode]
```

```python
new_iowa_Welfare_Benefit_Life_under_200['LIFE']=(new_iowa_Welfare_Benefit_Life_under_200.apply (lambda row: getBenefit (row,'Life'),axis=1))
```

```python
new_iowa_Welfare_Benefit_Life_under_200['DENTAL']=(new_iowa_Welfare_Benefit_Life_under_200.apply (lambda row: getBenefit (row,'Dental'),axis=1))
new_iowa_Welfare_Benefit_Life_under_200['VISION']=(new_iowa_Welfare_Benefit_Life_under_200.apply (lambda row: getBenefit (row,'Vision'),axis=1))
new_iowa_Welfare_Benefit_Life_under_200['TEMPORARY DISABILITY']=(new_iowa_Welfare_Benefit_Life_under_200.apply (lambda row: getBenefit (row,'Temporary disability'),axis=1))
new_iowa_Welfare_Benefit_Life_under_200['LONG-TERM DISABILITY']=(new_iowa_Welfare_Benefit_Life_under_200.apply (lambda row: getBenefit (row,'Long-term disability'),axis=1))
```
